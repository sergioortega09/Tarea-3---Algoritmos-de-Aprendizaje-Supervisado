```python
# 4) Dividir el dataset en Train y Test para evaluar correctamente el modelo

# Porcentaje de entrenamiento (train) = 80%
# Porcentaje de entrenamiento (test) = 20%

# Definir variables A y B


from sklearn.model_selection import train_test_split
Anexo4 = pd.read_csv('C:/Users/HP 15-gw0005la/Documents/ACTIVIDADES UNAD/Periodo 16-01-2024/Analisis de Datos/Tarea 3 - Algoritmos de Aprendizaje Supervisado/Anexo 4 - Dataset Red Wine Quality/calidadvino-rojo.csv')

A = Anexo4.drop("alcohol",axis=1)
B = Anexo4["alcohol"]

A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.20, random_state=0)
```
