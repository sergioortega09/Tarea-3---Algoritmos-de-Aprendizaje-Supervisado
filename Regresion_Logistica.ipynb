```python
import pandas as pd

Anexo3 = pd.read_csv('C:/Users/HP 15-gw0005la/Documents/ACTIVIDADES UNAD/Periodo 16-01-2024/Analisis de Datos/Tarea 3 - Algoritmos de Aprendizaje Supervisado/Anexo 3 - Dataset Heart Disease Cleveland UCI/heart_cleveland_upload.csv')

Anexo3["condicion"].value_counts()
```

```python
# 4) Dividir el dataset en Train y Test para evaluar correctamente el modelo

# Porcentaje de entrenamiento (train) = 80%
# Porcentaje de entrenamiento (test) = 20%

# Definir variables A y B


from sklearn.model_selection import train_test_split
Anexo3 = pd.read_csv('C:/Users/HP 15-gw0005la/Documents/ACTIVIDADES UNAD/Periodo 16-01-2024/Analisis de Datos/Tarea 3 - Algoritmos de Aprendizaje Supervisado/Anexo 3 - Dataset Heart Disease Cleveland UCI/heart_cleveland_upload.csv')

A = Anexo3.drop("condicion",axis=1)
B = Anexo3["condicion"]

A_train, A_test, B_train, B_test = train_test_split(A, B, test_size=0.20, random_state=0)
```

```python
print (A_train)
```

```python
print (B_train)
```

```python
# 5) Entrenar el modelo configurando los diferentes hiperparámetros.

# Importar la biblioteca necesaria
from sklearn.linear_model import LogisticRegression

regresion_logistica = LogisticRegression(solver = "liblinear")
regresion_logistica.fit(A_train , B_train)

Prediccion= regresion_logistica.predict(A_test)
```

```python
print (B_test)
```

```python
print (Prediccion)
```

```python
# 6) Evaluar el desempeño del modelo en el conjunto de Test con métricas como precisión, recall, F1-score, etc.
from sklearn.metrics import classification_report
print(classification_report(B_test,Prediccion))
```

```python
# 7)Realizar las diferentes gráficas que permitan visualizar los resultados del modelo.
# Se realiza la matriz de confusion del modelo

from sklearn import metrics
from sklearn.metrics import ConfusionMatrixDisplay
from sklearn.metrics import confusion_matrix

cm=confusion_matrix(B_test,Prediccion)
disp=metrics.ConfusionMatrixDisplay(confusion_matrix=cm)
disp.plot()
```

```python
# Curva de precision del modelo
from sklearn.metrics import RocCurveDisplay

RocCurveDisplay.from_predictions(B_test, Prediccion)
plt.show()
```
